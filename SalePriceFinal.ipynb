1. Data Cleaning

```python
df.isnull().sum()
#it seems there is no null value however when we check column values,we have null values.
```

```python
#check column values
for ex in df.columns:
 print(df[ex].value_counts())
  
```

```python
#delete column with no values
df.drop(['Unnamed: 0','APARTMENT NUMBER','EASE-MENT'],axis=1, inplace=True)
df
```

```python
#Delete Sale Price 0 and - values
df=df[df['SALE PRICE'] !='0']
df=df[df['SALE PRICE'] != ' -  ']
df['SALE PRICE']=df['SALE PRICE'].astype(int)
```

```python
#replace - for 0 first and 0  for mean values for 'LAND SQUARE FEET','GROSS SQUARE FEET'. We prefer not to delete rows with 0 and '-' to keep more data
ex1=['LAND SQUARE FEET','GROSS SQUARE FEET']
for x in ex1:
  df[x]=df[x].replace(' -  ',0)
  df[x]=df[x].astype(int)
  vl=int (df[x].mean())
  df[x]=df[x].replace(0,vl)
df.describe()
```

```python
#Create Age column
df['Age']=pd.DatetimeIndex(df['SALE DATE']).year-df['YEAR BUILT']
df
```

```python
#delete outlier if row values greater than absolute value of 3 for non categorical columns
from scipy import stats
ex2=['LAND SQUARE FEET','GROSS SQUARE FEET','SALE PRICE','Age','RESIDENTIAL UNITS','COMMERCIAL UNITS']
for x in ex2: 
  x2='z'+x
  df[x2]=stats.zscore(df[x])
  df = df[abs(df[x2])<3]
df
```

```python
#delete z scores. I deleted  SALE DATE','YEAR BUILT' as it is used in Age . 
ex3=['zAge','zLAND SQUARE FEET',  'zGROSS SQUARE FEET', 'zSALE PRICE', 'zRESIDENTIAL UNITS', 'zCOMMERCIAL UNITS','SALE DATE','YEAR BUILT']
df.drop(ex3,inplace=True,axis=1)
df
```

2. SQL Queries

```python
#SQL Queries for 2a
import numpy as np
import pandas as pd
import math
from sqlalchemy import create_engine
engine = create_engine('sqlite://')
raw_con = engine.raw_connection()
raw_con.create_function("sqrt", 1, math.sqrt)
df.to_sql("df", con=engine, if_exists='replace')
strSQL = """
with sales_stats as
(select avg([SALE PRICE]) as mean, sqrt((count([SALE PRICE])*sum([SALE PRICE]*[SALE PRICE]) - (sum([SALE PRICE])*sum([SALE PRICE])))/((count(*)-1)*(count([SALE PRICE])))) as stdev 
FROM df )
select 
   (df.[SALE PRICE] - sales_stats.mean) / sales_stats.stdev as sale_price_zscore 
 
from sales_stats,df ;"""

df2a = pd.read_sql(strSQL, engine)
df2a
```

```python
#SQL Queries for 2b. I can calculate z-score for 2a. However, I could not calculate standard deviation for 2b. for this reason, I calculate square of z-score
import numpy as np
import pandas as pd
import math
from sqlalchemy import create_engine
engine = create_engine('sqlite://')
raw_con = engine.raw_connection()
raw_con.create_function("sqrt", 1, math.sqrt)
#raw_con.create_function("stdev", 1, np.mean())
df.to_sql("df", con=engine, if_exists='replace')
strSQL = """
with sales_stats as(
select NEIGHBORHOOD,[BUILDING CLASS CATEGORY],avg([SALE PRICE]) as mean,(count([SALE PRICE])*sum([SALE PRICE]*[SALE PRICE]) - (sum([SALE PRICE])*sum([SALE PRICE])))/((count(*)-1)*(count([SALE PRICE]))) as variance 
FROM df group by NEIGHBORHOOD,[BUILDING CLASS CATEGORY])
select 
   df.NEIGHBORHOOD,df.[BUILDING CLASS CATEGORY], (df.[SALE PRICE] - sales_stats.mean)* (df.[SALE PRICE] - sales_stats.mean)/ sales_stats.variance as sale_price_zscore2 
 
from sales_stats,df where sales_stats.NEIGHBORHOOD=df.NEIGHBORHOOD and sales_stats.[BUILDING CLASS CATEGORY]=df.[BUILDING CLASS CATEGORY] ;
"""

df2b = pd.read_sql(strSQL, engine)
df2b
```

3. Model and Analysis- 

```python
 #I deleted Total units since there is high correlation between total and residential units.
df.drop('TOTAL UNITS',inplace=True,axis=1)
df.columns
```

```python
#change categorical variable to category type as they have categorical values.
ex4=['BOROUGH','ZIP CODE','TAX CLASS AT TIME OF SALE']
for x in ex4:
 df[x]=df[x].astype('category')
df.dtypes
```

```python
#change object type to category type
for c in df.columns:
    col_type = df[c].dtype
    if col_type == 'object' or col_type.name == 'category':
        df[c] = df[c].astype('category')
df.dtypes
```

```python
#Train test split
df3=df.copy()
y=df3['SALE PRICE']
df3=df3.drop('SALE PRICE',axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df3, y, test_size=0.33)
```

```python
df3.dtypes
```

```python
#Identify categorical variables
categorical_features_indices = np.where(df3.dtypes == 'category')[0]

categorical_features_indices
```

```python
# I will use catboost since it can handle categorical variables without changing categorical variables into dummy variables
!pip install catboost
```

```python
# I will use catboost since it can handle categorical variables without changing categorical variables into dummy variables

import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
model=CatBoostRegressor(iterations=500, depth=3, learning_rate=0.1, loss_function='RMSE')
model.fit(X_train, y_train,cat_features=categorical_features_indices,eval_set=(X_test, y_test),verbose=False)
```

```python
#To get the most important features, I used feature get_feature_importance() of CatBoost
feat_imp = pd.Series(model.get_feature_importance(), index=df3.columns)
feat_imp.nlargest(30).plot(kind='barh', figsize=(8,10))
#It seems that size, location and age is important for Sale Price
```
